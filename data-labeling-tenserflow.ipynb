# Code Cell: Import the necessary libraries

```python
# Step 2: Import the required libraries
import os
import shutil
import numpy as np
import pandas as pd
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2, preprocess_input, decode_predictions
from tensorflow.keras.preprocessing.image import load_img, img_to_array
```

# Code Cell: Load the pre-trained MobileNetV2 model.

```python
# Step 3: Load the pre-trained MobileNetV2 model
model = MobileNetV2(weights='imagenet')
```

# Code Cell: Label the images and move them to the corresponding folders.

```python
# Step 4: Label the images and move them to corresponding folders
input_folder_path = '/kaggle/input/sample-images-for-kaggle-demos'
output_folder_path = '/kaggle/working'

# Function to label images and move them to corresponding folders
def label_and_move_images(folder_path):
    for filename in os.listdir(folder_path):
        image_path = os.path.join(folder_path, filename)
        image = load_img(image_path, target_size=(224, 224))
        image = img_to_array(image)
        image = preprocess_input(image)
        image = np.expand_dims(image, axis=0)

        prediction = model.predict(image)
        predicted_label = decode_predictions(prediction, top=1)[0][0][1]

        # Create a folder with the predicted label and move the image into it
        label_folder = os.path.join(output_folder_path, predicted_label)
        os.makedirs(label_folder, exist_ok=True)
        destination_path = os.path.join(label_folder, filename)
        shutil.copy(image_path, destination_path)

# Call the function to label images and move them
label_and_move_images(input_folder_path)

print("Image labeling and organization completed.")
```
